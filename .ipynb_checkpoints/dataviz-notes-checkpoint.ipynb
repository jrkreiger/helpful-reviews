The purpose of this notebook is to create/clean the data for a Tableau story on "helpful" reviews.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('./data/Reviews.csv')
data.head()
```

```python
# Drop duplicate reviews based on identical texts
data_no_dupe_texts = data.drop_duplicates(subset=['Text'])
```

```python
# Create Text_len column
data_no_dupe_texts['Text_len'] = [len(data_no_dupe_texts.loc[row, 'Text']) for row in data_no_dupe_texts.index]
```

```python
# Get average length of a review that only got downvotes on helpfulness
data_no_dupe_texts[(data_no_dupe_texts.HelpfulnessDenominator > 0) & (data_no_dupe_texts.HelpfulnessNumerator == 0)]['Text_len'].mean()
```

```python
# Get average length of a review with no votes either way
data_no_dupe_texts[(data_no_dupe_texts.HelpfulnessDenominator == 0) & (data_no_dupe_texts.HelpfulnessNumerator == 0)]['Text_len'].mean()
```

```python
# Get ProductId value counts
prod_id_counts = data_no_dupe_texts.ProductId.value_counts()
```

```python
# Make column for ProductIdCount
data_no_dupe_texts['ProductIdCount'] = [prod_id_counts[idx] for idx in data_no_dupe_texts['ProductId']]
```

```python
# Make column for ProductMostReviewed
data_no_dupe_texts['ProductMostReviewed'] = np.where(data_no_dupe_texts['ProductIdCount'] > np.quantile(prod_id_counts, 0.99), 1, 0)
```

```python
# Make column for UserIdCount
user_id_counts = data_no_dupe_texts.UserId.value_counts()
data_no_dupe_texts['UserIdCount'] = [user_id_counts[idx] for idx in data_no_dupe_texts['UserId']]
```

```python
# Make column for TopReviewer
data_no_dupe_texts['TopReviewer'] = np.where(data_no_dupe_texts['UserIdCount'] > np.quantile(user_id_counts, 0.99), 1, 0)
```

```python
data_no_dupe_texts['Downvotes'] = data_no_dupe_texts['HelpfulnessDenominator'] - data_no_dupe_texts['HelpfulnessNumerator']
```

```python
# Define a function to run Welch's t-test, evaluate and print results
def run_welch(a, b, two_sided=False, alpha=0.05):
    """Perform Welch's t-test and compare p-value to given alpha.
       a: first sample (one-dimensional array)
       b: second sample (one-dimensional array)
       two-sided: whether the t-test should be two-sided or not (boolean)
       alpha: alpha value for comparison with calculated p-value
       
       Dependencies: welch_functions.py"""
    
    # Calculate p-value
    p_value = p_value_welch_ttest(a, b, two_sided=two_sided)
    
    # Evaluate and print results
    if p_value < alpha:
        print('Null hypothesis rejected (p_value = {})'.format(p_value))
    else:
        print('Fail to reject null hypothesis (p_value = {})'.format(p_value)) 
        

# Define a function to calculate Cohen's d
def cohen_d(sample_a, sample_b):
    """Calculate's Cohen's d for two 1-D arrays"""
    
    diff = abs(np.mean(sample_a) - np.mean(sample_b))
    n1 = len(sample_a) 
    n2 = len(sample_b)
    var1 = np.var(sample_a)
    var2 = np.var(sample_b)
    pooled_var = (n1 * var1 + n2 * var2) / (n1 + n2)
    
    d = diff / np.sqrt(pooled_var)
    return d

# Define a function to calculate effect size and power of a statistical test
def size_and_power(sample_a, sample_b, alpha=0.05):
    """Prints raw effect size and power of a statistical test, using 
       Cohen's d (calculated with Satterthwaite approximation)
       Dependencies: Numpy, statsmodels.stats.TTestIndPower"""
    
    effect = abs(np.mean(sample_a) - np.mean(sample_b))
    print('Raw effect size:', effect)

    d = cohen_d(sample_a, sample_b)
    print('Cohen\'s d:', d)
    power_analysis = TTestIndPower()
    
    sample_length = (len(sample_a) + len(sample_b)) / 2
    power = power_analysis.solve_power(effect_size=d, alpha=alpha, nobs1=sample_length)
        
    print('Power:', power)
```
